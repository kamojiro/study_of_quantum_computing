```python
import chainer
import chainer.functions as F
import chainer.links as L
from chainer import training
from chainer.training import extensions
import math
```

```python
train, test = chainer.datasets.get_mnist(ndim=3)
```

```python
gpu = 0         #使用するgpuの番号を指定．gpu=-1でCPUを使うことになる．
unit = 100      #モデルの中間層の横幅を指定．
batchsize = 100 #ミニバッチサイズ．
epoch = 20      #学習のエポック数を指定．（データを何ループ見るか）
out = 'result'  #結果の出力ディレクトリ
plot = True     #途中結果をグラフに図示するか．

print('# GPU: {}'.format(gpu))
print('# unit: {}'.format(unit))
print('# Minibatch-size: {}'.format(batchsize))
print('# epoch: {}'.format(epoch))
print('# output directry: {}'.format(out))
print('# plot intermediate results: {}'.format(plot))
print('')

resume = False  #学習途中から再開するか
```

```python
class MyCNN(chainer.Chain):
    def __init__(self, n_units, n_out, stride=1):
        w = math.sqrt(2)
        super(MyCNN, self).__init__(
            conv1 = L.Convolution2D(None, n_units, 4, stride, 1, w),
            conv2 = L.Convolution2D(None, n_units, 4, stride, 1, w),
            conv3 = L.Convolution2D(None, n_units, 4, stride, 1, w),
            conv4 = L.Convolution2D(None, n_units, 4, stride, 1, w),
            l1 = L.Linear(None, n_out)
        )
        self.train = True

    def __call__(self, x):
        h1 = F.relu(self.conv1(x))
        h2 = F.relu(self.conv2(h1))
        h3 = F.relu(self.conv3(h2))
        h4 = F.relu(self.conv4(h3))
        return self.l1(h4)  #最終層は全結合層
model = L.Classifier(MyCNN(unit, 10, stride=1)) #modelというオブジェクトにMLPモデルを格納．出力は10クラスの分類問題．
if gpu >= 0:
    # Make a specified GPU current
    chainer.cuda.get_device(gpu).use() 
    model.to_gpu()  # Copy the model to the GPU
```

```python
optimizer = chainer.optimizers.Adam()
optimizer.setup(model)
```

```python
# 学習の反復を行うiteratorの準備
train_iter = chainer.iterators.SerialIterator(train, batchsize)
test_iter = chainer.iterators.SerialIterator(test, batchsize,
                                             repeat=False, shuffle=False)

# Set up a trainer
updater = training.updaters.StandardUpdater(train_iter, optimizer, device=gpu)
trainer = training.Trainer(updater, (epoch, 'epoch'), out=out)
```

```python
# Evaluate the model with the test dataset for each epoch
trainer.extend(extensions.Evaluator(test_iter, model, device=gpu))

# Dump a computational graph from 'loss' variable at the first iteration
# The "main" refers to the target link of the "main" optimizer.
trainer.extend(extensions.dump_graph('main/loss'))

# Take a snapshot for each specified epoch
frequency = 1
trainer.extend(extensions.snapshot(), trigger=(frequency, 'epoch'))

# Write a log of evaluation statistics for each epoch
trainer.extend(extensions.LogReport())

# Save two plot images to the result dir
# 学習の経過は./result/loss.pngや./result/accuracy.pngに出力される．
if plot and extensions.PlotReport.available():
    trainer.extend(
        extensions.PlotReport(['main/loss', 'validation/main/loss'],
                              'epoch', file_name='loss.png'))
    trainer.extend(
        extensions.PlotReport(
            ['main/accuracy', 'validation/main/accuracy'],
            'epoch', file_name='accuracy.png'))

# Print selected entries of the log to stdout
# Here "main" refers to the target link of the "main" optimizer again, and
# "validation" refers to the default name of the Evaluator extension.
# Entries other than 'epoch' are reported by the Classifier link, called by
# either the updater or the evaluator.
# ./result/logに途中経過のログが出力される．
trainer.extend(extensions.PrintReport(
    ['epoch', 'main/loss', 'validation/main/loss',
     'main/accuracy', 'validation/main/accuracy', 'elapsed_time']))

# Print a progress bar to stdout
# プログレスバーを表示する．
trainer.extend(extensions.ProgressBar())

if resume:  #途中から再開する場合．
  # Resume from a snapshot
  chainer.serializers.load_npz(resume, trainer)
```

```python
trainer.run()
```

```python
import pandas as pd
import json
with open('./result/log') as f:
    logs = json.load(f)
    results = pd.DataFrame(logs)
results[['main/accuracy', 'validation/main/accuracy']].plot()
```
