```python
import chainer
from chainer import functions as F
from chainer import links as L
from chainer import training
from chainer.training import extensions


train, test = chainer.datasets.get_cifar10()

class Block(chainer.Chain):

    def __init__(self, out_channels, ksize, pad=1):
        super(Block, self).__init__()
        with self.init_scope():
            self.conv = L.Convolution2D(None, out_channels, ksize, pad=pad,
                                        nobias=True)
            self.bn = L.BatchNormalization(out_channels)

    def forward(self, x):
        h = self.conv(x)
        h = self.bn(h)
        return F.relu(h)
      
class smallVGG(chainer.Chain):

    def __init__(self, class_labels=10):
        super(smallVGG, self).__init__()
        with self.init_scope():
            self.block1_1 = Block(64, 3)
            self.block1_2 = Block(64, 3)
            self.block2_1 = Block(128, 3)
            self.block2_2 = Block(128, 3)
            self.block3_1 = Block(256, 3)
            self.block3_2 = Block(256, 3)
            self.block3_3 = Block(256, 3)
            self.block3_4 = Block(256, 3)
            self.fc1 = L.Linear(None, 512, nobias=True)
            self.bn_fc1 = L.BatchNormalization(512)
            self.fc2 = L.Linear(None, class_labels, nobias=True)

    def forward(self, x):
        # 64 channel blocks:
        h = self.block1_1(x)
        h = self.block1_2(h)
        h = F.max_pooling_2d(h, ksize=2, stride=2)
        h = F.dropout(h, ratio=0.25)
        
        # 128 channel blocks:
        h = self.block2_1(h)
        h = self.block2_2(h)
        h = F.max_pooling_2d(h, ksize=2, stride=2)
        h = F.dropout(h, ratio=0.25)

        # 256 channel blocks:
        h = self.block3_1(h)
        h = self.block3_2(h)
        h = self.block3_3(h)
        h = self.block3_4(h)
        h = F.max_pooling_2d(h, ksize=2, stride=2)
        h = F.dropout(h, ratio=0.25)

        h = self.fc1(h)
        h = self.bn_fc1(h)
        h = F.relu(h)
        h = F.dropout(h, ratio=0.5)
        return self.fc2(h)
      

batchsize = 100
train_iter = chainer.iterators.SerialIterator(train, batchsize)
test_iter = chainer.iterators.SerialIterator(test, batchsize,
                                             repeat=False, shuffle=False)

model = L.Classifier(smallVGG())

gpu = 0
if gpu >= 0:
    chainer.cuda.get_device(gpu).use() 
    model.to_gpu()  

opt = chainer.optimizers.Adam()
opt.setup(model)

epoch = 15

updater = training.StandardUpdater(train_iter, opt, device=gpu)
trainer = training.Trainer(updater, (epoch, 'epoch'), out='/cifar/result')

trainer.extend(extensions.Evaluator(test_iter, model, device=gpu))

trainer.extend(extensions.dump_graph('main/loss'))

trainer.extend(extensions.snapshot(), trigger=(epoch, 'epoch'))

trainer.extend(extensions.LogReport())

trainer.extend(extensions.PrintReport(
    ['epoch', 'main/loss', 'validation/main/loss',
     'main/accuracy', 'validation/main/accuracy']))

trainer.extend(extensions.ProgressBar())

trainer.run()

print('save the model')
chainer.serializers.save_hdf5('VGG_CIFAR.model', model)
print('save the optimizer')
chainer.serializers.save_hdf5('VGG_CIFAR.state', opt)
```

```python
import pandas as pd
import json
with open('/cifar/result/log') as f:
    logs = json.load(f)
    results = pd.DataFrame(logs)
results[['main/accuracy', 'validation/main/accuracy']].plot()
```
