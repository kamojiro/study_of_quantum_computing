```python
import numpy as np
from sklearn import datasets
from sklearn import linear_model
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import sklearn.metrics as skmet
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
from PIL import Image
```

```python
#bostonのデータ
boston = datasets.load_boston()
X = boston.data
```

```python
def ridge(lam, features, target):#features=13, target = 506
    l, r = features.shape
    lameye = np.eye(r+1)*lam
    Ivec = np.array(([[1] for _ in range(l)]))
    X = np.concatenate((Ivec, features), axis = 1) #定数部分に相当するものを付け加える
    #(506,14)になる
    a = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)+ lameye), X.T),target)
    #(X.T*X + lam*I)^{-1}X.Ty
    return np.dot(a.T, X.T)
    
```

```python
predicted = ridge(.5, boston.data, boston.target)
testy = boston.target
```

```python
fig, ax = plt.subplots()
ax.scatter(testy, predicted)
ax.plot([testy.min(), testy.max()], [testy.min(), testy.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
print(np.sum( (predicted - testy)**2) + 0.5*np.sum(predicted**2))
```

```python
clf = linear_model.Ridge (alpha = .5)
clf.fit(boston.data, boston.target)
predicted_sklearn = clf.predict(boston.data)
print(mean_squared_error(predicted_sklearn, ridge(0.5, boston.data, boston.target)))
```

リッジ回帰を実装し，その結果とscikit-learnのリッジ回帰を用いた結果の平均2乗和を取ると，上の値が出た．若干のズレがあるが，逆行列を取る際の計算に原因があると考えられる．
